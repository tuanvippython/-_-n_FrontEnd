```python
# Danh sách các cột
columns = ["Price", "Production_year", "Mileage_km", "Power_HP", "Displacement_cm3", "CO2_emissions"]

# Số lượng cột và dòng trong biểu đồ chung
n_cols = 2
n_rows = 3

# Tạo subplot với 4 dòng và 2 cột
fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(15, 15))

# Duyệt qua từng cột và vẽ biểu đồ mật độ
for i, col in enumerate(columns):
    row_index = i // n_cols
    col_index = i % n_cols
    # Vẽ biểu đồ mật độ
    sns.kdeplot(df[col], fill=True, color='orange', ax=axes[row_index, col_index])
    # Thiết lập tiêu đề và nhãn
    axes[row_index, col_index].set_title(f'Density Plot of {col}',fontsize = 14)
    axes[row_index, col_index].set_xlabel(col)
    # Chỉnh kích thước của nhãn trục x và y
    axes[row_index, col_index].set_xlabel(col, fontsize=14)
    axes[row_index, col_index].set_ylabel('Density', fontsize=14)
     # Điều chỉnh thông số của trục x và y
    axes[row_index, col_index].tick_params(axis='x', labelsize=12)
    axes[row_index, col_index].tick_params(axis='y', labelsize=12)

# Tinh chỉnh layout
plt.tight_layout()

# Hiển thị biểu đồ chung
plt.show()
```

```python
# Chọn các cột cần vẽ biểu đồ tròn
selected_columns = ["Currency", "Condition", "Fuel_type", "Drive", "Transmission", "Type", "Colour", "Doors_number"]

# Tạo figure và axes
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(15, 20))
fig.suptitle('Distribution of Various Features', fontsize=16)

# Duyệt qua từng cột và vẽ biểu đồ tròn
for i, col in enumerate(selected_columns):
    # Tính số lượng giá trị khác nhau trong cột
    value_counts = df[col].value_counts()

    # Vẽ biểu đồ tròn trên trục tương ứng
    ax = axes[i // 2, i % 2]
    wedges, texts, autotexts = ax.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', colors=plt.cm.Paired.colors)
    ax.set_title(f'Distribution of {col}')
# Tinh chỉnh layout
plt.tight_layout(rect=[0, 0, 1, 0.96])

# Hiển thị biểu đồ
plt.show()
```

```python
# Bỏ những cột chứa nhiều giá trị thiếu và những cột không cần thiết trong bài toán
df =df.drop(["Index",'Vehicle_version', 'Vehicle_generation','CO2_emissions','Origin_country'
             ,'First_owner','First_registration_date',"Features","Offer_location","Offer_publication_date"], axis=1)
df.info
```

```python
# Xem các giá trị thiếu
df.isnull().sum()
```

```python
# Duyệt qua những cột có chứa giá trị thiếu và điền giá trị nhiều nhất của mỗi cột vào dữ liệu thiếu
for column in df.columns[df.isnull().any()]:
    gtri_xh_nhieu_nhat = df[column].value_counts().idxmax()
    print(f"Cột: {column}")
    print(f"Giá trị xuất hiện nhiều nhất: {gtri_xh_nhieu_nhat}")
    df[column].fillna(gtri_xh_nhieu_nhat, inplace=True)
```

```python
df.isnull().sum()
```

```python
# Kiểm tra các giá trị khác nhau của mỗi cột
for column in df.select_dtypes(include='object').columns:
    print(f"{column}: {df[column].unique()}")
```

```python
# Mã hóa dữ liệu
from sklearn.preprocessing import LabelEncoder
# Các cột cần mã hóa
select_columns = ["Condition", "Fuel_type", "Type"]
# Khởi tạo LabelEncoder object
label_encoder = LabelEncoder()
# Áp dụng LabelEncoder cho từng cột
for column in select_columns:
    # Mã hóa chuỗi thành số nguyên và gán kết quả vào cột cũ
    df[column] = label_encoder.fit_transform(df[column])
# Hiển thị DataFrame sau khi đã mã hóa
df
```

```python
# Mã hóa dữ liệu
for columns_to_encode in df.select_dtypes(include='object').columns:
    # Mã hóa các cột bằng dummy variables và chuyển đổi thành giá trị 0 và 1
    dummies = pd.get_dummies(df[columns_to_encode], prefix=columns_to_encode, drop_first=True).astype(int)
    # Thêm các cột đã mã hóa vào DataFrame gốc
    df = pd.concat([df, dummies], axis=1)
    # Loại bỏ các cột gốc đã được mã hóa
    df = df.drop(columns=columns_to_encode)
# Hiển thị DataFrame sau khi thực hiện
df
```

```python
# Chuẩn hóa dữ liệu
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# Chọn các cột cần chuẩn hóa
columns = ["Price", "Production_year", "Mileage_km","Power_HP","Displacement_cm3","Doors_number"]
# Chuẩn hóa dữ liệu và gán lại cho DataFrame
df[columns] = scaler.fit_transform(df[columns])
# In DataFrame sau khi chuẩn hóa
df
```

```python
# Xử lí các Outliers
from scipy.stats import zscore
# Chọn các cột cần xử lý
select_columns = df[['Price', 'Production_year','Mileage_km',"Power_HP","Displacement_cm3"]]
# Tính z-scores cho từng cột riêng lẻ
z_scores = zscore(select_columns, axis=0)
# Áp dụng điều kiện cho từng cột riêng lẻ
df = df[(abs(z_scores) < 3).all(axis=1)]
# In ra DataFrame sau khi loại bỏ outliers
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Biểu đồ boxplot trước khi xử lý outliers
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
sns.boxplot(data=select_columns)
plt.title('Dữ liệu trước khi xử lí Outliers')
# Biểu đồ boxplot sau khi xử lý outliers
plt.subplot(2, 1, 2)
sns.boxplot(data=df[['Price', 'Production_year', 'Mileage_km', 'Power_HP', 'Displacement_cm3']])
plt.title('Dữ liệu sau khi xử lí Outliers')
plt.tight_layout()
plt.show()
```
